## §2 · Componente Chains + LCEL

**🎯 Objetivo:** mostrar la pieza mínima de LangChain — la cadena `prompt | modelo | parser` —
antes de que quede embebida dentro del RAG y del agente. La probamos con los **conceptos
centrales del capítulo**, no con ejemplos genéricos.

**⚙️ Qué hace el bloque:** define la chain con LCEL y la ejecuta de las tres maneras:
`invoke` (una llamada), `batch` (varias en paralelo) y `stream` (token a token, que es
lo que después usa la interfaz).

**🗣️ Al presentar:** "esta barra vertical es LCEL: compone prompt, modelo y parser en un
pipeline. Todo lo que sigue en el proyecto son variaciones de esta misma idea".

```python
# ============================================================
# ⛓️ §2 · Chain LCEL: invoke / batch / stream
# ============================================================
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

llm    = ChatOpenAI(model=MODEL_NAME, temperature=0.2)
parser = StrOutputParser()

demo_prompt = ChatPromptTemplate.from_template(
    "En una oración, para un decisor de política pública, explicá qué significa: {concepto}"
)
demo_chain = demo_prompt | llm | parser          # ← LCEL

print("invoke →", demo_chain.invoke({"concepto": "accesibilidad comunicacional"}))

for r in demo_chain.batch([{"concepto": "ajustes razonables"},
                           {"concepto": "gobernanza de datos"}]):
    print("batch  →", r)

print("stream →", end=" ")
for chunk in demo_chain.stream({"concepto": "tecnologías de apoyo"}):
    print(chunk, end="")
print()
```

```python
# ============================================================
# 📚 §3.1 · Cargar el capítulo (upload .docx o resumen real)
# ============================================================
from pathlib import Path

DOCS_DIR = Path("/content/docs_bid"); DOCS_DIR.mkdir(exist_ok=True)

subidos = []
try:
    from google.colab import files
    print("📤 Subí el capítulo (.docx / .pdf / .md) o Cancelá para usar el resumen…")
    up = files.upload()
    for nombre, contenido in up.items():
        destino = DOCS_DIR / nombre
        destino.write_bytes(contenido)
        subidos.append(destino)
        print("   ✔ guardado:", destino)
except Exception:
    pass

if not subidos:
    # --- Resumen FIEL del capítulo (datos reales de las Tablas 1, 2 y 3) ---
    demo = DOCS_DIR / "resumen_capitulo_bid.md"
    demo.write_text("""# IA accesible e inclusión laboral de personas con discapacidad auditiva en Argentina
Autor: Julio Alberto Paredes Rojas (Pontificia Universidad Católica Argentina).
Capítulo del libro Uniandes–BID "Inteligencia artificial en el sector público de América Latina y el Caribe".

## Pregunta que orienta el capítulo
¿Qué lecciones aporta la experiencia de las personas con discapacidad auditiva en Argentina
para construir una hoja de ruta pública de IA accesible? El caso analizado no es un sistema
estatal único, sino una condición habilitante: un marco de gobernanza que permita incorporar
tecnologías accesibles a las políticas de empleo, formación y protección de derechos.

## Metodología (Tabla 1)
- Encuesta digital: 170 participantes (tendencias sobre empleo, barreras, tecnología, formación y privacidad).
- Narrativas laborales: 50 testimonios (experiencias, obstáculos, estrategias de adaptación).
- Entrevistas institucionales: Dillo.ai, Eldes y Sumate (modelos de intervención y lecciones organizacionales).
- Análisis documental: normas, informes y literatura.
- Procesamiento analítico: R, Shiny, minería de texto y Random Forest.

## Hallazgos clave (Tabla 2)
- Uso de tecnologías de apoyo: 70 % utiliza alguna; 30 % no utiliza → se requiere acceso universal, orientación y apoyo técnico.
- Rotación laboral: 56,7 % tuvo dos o más empleos en cinco años → la inclusión debe medir continuidad, no solo contratación.
- Sin inserción laboral: cerca del 10 % → persisten barreras estructurales de acceso.
- Principal barrera: 33,3 % mencionó falta de adaptaciones tecnológicas → los ajustes deben institucionalizarse y financiarse.
- Apoyo de ONG o entidades: 60 % recibió apoyo → las organizaciones intermedias cubren vacíos del Estado.
- Prioridad asignada a la IA: cerca del 50 % alta, 30 % media, 20 % baja → la adopción debe responder a necesidades y confianza, no imponerse.
- Privacidad: 68,3 % asignó máxima importancia → la gobernanza de datos es condición de legitimidad.

## Hoja de ruta pública para IA accesible (Tabla 3, siete etapas)
1. Diagnóstico — Mapear barreras, usuarios, procesos y tecnologías existentes. Producto verificable: línea de base e inventario de necesidades.
2. Cocreación — Formar un comité con comunidad sorda, Estado, academia y proveedores. Producto: requisitos funcionales y de derechos.
3. Estándares — Definir criterios de accesibilidad, privacidad, seguridad e interoperabilidad. Producto: protocolo técnico y ético.
4. Piloto — Probar herramientas en procesos concretos y con usuarios diversos. Producto: informe de desempeño e impacto.
5. Formación — Capacitar usuarios, funcionarios, compras, RRHH y soporte técnico. Producto: plan formativo accesible y evaluado.
6. Escalamiento — Financiar e integrar soluciones que demuestren utilidad. Producto: programa institucional con presupuesto.
7. Evaluación y rendición de cuentas — Medir resultados, incidentes, reclamos y desigualdades. Producto: tablero público e informes periódicos.

## Idea central
El impacto de las tecnologías depende menos de su novedad que de las condiciones
institucionales que permiten adoptarlas, financiarlas, supervisarlas y usarlas de manera
sostenible. La accesibilidad se desplaza del dispositivo hacia la capacidad pública,
en línea con una concepción de igualdad efectiva de oportunidades (Rawls, 1971) y con
los marcos de OMS (2021) y UNESCO (2021).
""", encoding="utf-8")
    subidos = [demo]
    print("📄 Usando resumen fiel del capítulo:", demo.name)
    print("   (para la demo completa, subí el .docx del capítulo)")

print(f"\n📁 Corpus del RAG: {[p.name for p in subidos]}")
```

```python
# ============================================================
# 📚 §3.2 · Cargar (.docx/.pdf/.md) + dividir en chunks
# ============================================================
from langchain_core.documents import Document
from langchain_text_splitters import RecursiveCharacterTextSplitter

def cargar_documentos(paths):
    """Carga cada archivo según su formato. El .docx se lee párrafo a
    párrafo y las TABLAS se convierten a texto (clave: en el capítulo,
    los indicadores y la hoja de ruta viven en tablas de Word)."""
    docs = []
    for p in paths:
        p = Path(p)
        suf = p.suffix.lower()
        if suf == ".pdf":
            from langchain_community.document_loaders import PyPDFLoader
            docs += PyPDFLoader(str(p)).load()
        elif suf == ".docx":
            from docx import Document as DocxDocument
            d = DocxDocument(str(p))
            partes = [par.text for par in d.paragraphs if par.text.strip()]
            for ti, tabla in enumerate(d.tables, start=1):
                filas = [" | ".join(c.text.strip() for c in fila.cells)
                         for fila in tabla.rows]
                partes.append(f"[Tabla {ti}]\n" + "\n".join(filas))
            docs.append(Document(page_content="\n\n".join(partes),
                                 metadata={"source": p.name}))
        else:
            docs.append(Document(page_content=p.read_text(encoding="utf-8", errors="ignore"),
                                 metadata={"source": p.name}))
    return docs

documentos = cargar_documentos(subidos)

splitter = RecursiveCharacterTextSplitter(
    chunk_size=700, chunk_overlap=100,
    separators=["\n## ", "\n### ", "\n\n", "\n", ". ", " "],
)
chunks = splitter.split_documents(documentos)
print(f"🔪 {len(documentos)} documento(s) → {len(chunks)} chunks")
print("--- chunk[0] ---"); print(chunks[0].page_content[:280])
```

```python
# ============================================================
# 📚 §3.3 · Chroma PERSISTENTE + retriever
# ============================================================
from langchain_openai import OpenAIEmbeddings
from langchain_chroma import Chroma

PERSIST_DIR = "/content/chroma_bid"
COLLECTION  = "capitulo_bid"

embeddings = OpenAIEmbeddings(model="text-embedding-3-small")

vectorstore = Chroma(collection_name=COLLECTION,
                     embedding_function=embeddings,
                     persist_directory=PERSIST_DIR)     # ← persistencia

if vectorstore._collection.count() == 0:
    vectorstore.add_documents(chunks)
    print(f"🧬 Base creada y persistida en {PERSIST_DIR} "
          f"({vectorstore._collection.count()} vectores)")
else:
    print(f"♻️  Base EXISTENTE en {PERSIST_DIR} "
          f"({vectorstore._collection.count()} vectores) — no se re-embeddea.")

retriever = vectorstore.as_retriever(search_kwargs={"k": 4})

for d in vectorstore.similarity_search("¿cuál fue la principal barrera reportada?", k=2):
    print("→", d.page_content[:110].replace("\n", " "), "…")
```

```python
# ============================================================
# 📚 §3.4 · RAG chain (LCEL) con respuesta + fuentes
# ============================================================
from langchain_core.runnables import RunnablePassthrough

rag_prompt = ChatPromptTemplate.from_template("""Sos un asesor de políticas públicas. Respondé SOLO con la información
del contexto (proviene del capítulo Uniandes–BID de Paredes Rojas sobre IA accesible).
Si algo no figura en el contexto, decilo con claridad. Citá porcentajes exactos cuando existan.

CONTEXTO:
{context}

PREGUNTA: {question}

Respuesta precisa, en tono profesional para un funcionario:""")

def _format_docs(docs):
    return "\n\n".join(d.page_content for d in docs)

rag_chain = (
    {"context": retriever | _format_docs, "question": RunnablePassthrough()}
    | rag_prompt | llm | parser
)

def responder_con_fuentes(pregunta: str) -> dict:
    docs = retriever.invoke(pregunta)
    respuesta = rag_chain.invoke(pregunta)
    fuentes = sorted({d.metadata.get("source", "?") for d in docs})
    return {"respuesta": respuesta, "fuentes": fuentes}

r = responder_con_fuentes("¿Qué porcentaje usa tecnologías de apoyo y cuál fue la principal barrera?")
print(r["respuesta"]); print("📎", r["fuentes"])
print()
print(rag_chain.invoke("¿Cuál es el PBI de Noruega?"))   # fuera del corpus → debe decir que no figura
```

3. **§4.3** — Convierte dinámicamente cada tool MCP en una `StructuredTool` de LangChain

adentro: los consulta, igual que consultaría un sistema estatal real".

```python
# ============================================================
# 🔌 §4.1 · MCP Server con la evidencia real del capítulo
# ============================================================
from typing import Any
from fastmcp import FastMCP

mcp = FastMCP(
    name="observatorio-ia-accesible",
    instructions=(
        "Observatorio de políticas públicas de IA accesible. Sirve la hoja de ruta "
        "de 7 etapas y los indicadores empíricos del capítulo Uniandes–BID "
        "(Paredes Rojas) sobre inclusión laboral de personas con discapacidad "
        "auditiva en Argentina. Los policy briefs son borradores y requieren "
        "aprobación humana."
    ),
)

# --- Tabla 3 del capítulo: hoja de ruta pública (datos reales) ---
HOJA_DE_RUTA: dict[int, dict[str, str]] = {
    1: {"etapa": "Diagnóstico",
        "accion": "Mapear barreras, usuarios, procesos y tecnologías existentes",
        "producto_verificable": "Línea de base e inventario de necesidades"},
    2: {"etapa": "Cocreación",
        "accion": "Formar un comité con comunidad sorda, Estado, academia y proveedores",
        "producto_verificable": "Requisitos funcionales y de derechos"},
    3: {"etapa": "Estándares",
        "accion": "Definir criterios de accesibilidad, privacidad, seguridad e interoperabilidad",
        "producto_verificable": "Protocolo técnico y ético"},
    4: {"etapa": "Piloto",
        "accion": "Probar herramientas en procesos concretos y con usuarios diversos",
        "producto_verificable": "Informe de desempeño e impacto"},
    5: {"etapa": "Formación",
        "accion": "Capacitar usuarios, funcionarios, compras, RRHH y soporte técnico",
        "producto_verificable": "Plan formativo accesible y evaluado"},
    6: {"etapa": "Escalamiento",
        "accion": "Financiar e integrar soluciones que demuestren utilidad",
        "producto_verificable": "Programa institucional con presupuesto"},
    7: {"etapa": "Evaluación y rendición de cuentas",
        "accion": "Medir resultados, incidentes, reclamos y desigualdades",
        "producto_verificable": "Tablero público e informes periódicos"},
}

# --- Tabla 2 del capítulo: indicadores empíricos (datos reales) ---
INDICADORES: list[dict[str, str]] = [
    {"indicador": "Uso de tecnologías de apoyo",
     "resultado": "70 % utiliza alguna; 30 % no utiliza",
     "implicacion": "Se requiere acceso universal, orientación y apoyo técnico"},
    {"indicador": "Rotación laboral",
     "resultado": "56,7 % tuvo dos o más empleos en cinco años",
     "implicacion": "La inclusión debe medir continuidad, no solo contratación"},
    {"indicador": "Sin inserción laboral",
     "resultado": "Cerca del 10 %",
     "implicacion": "Persisten barreras estructurales de acceso"},
    {"indicador": "Principal barrera",
     "resultado": "33,3 % mencionó falta de adaptaciones tecnológicas",
     "implicacion": "Los ajustes deben institucionalizarse y financiarse"},
    {"indicador": "Apoyo de ONG o entidades",
     "resultado": "60 % recibió apoyo",
     "implicacion": "Las organizaciones intermedias cubren vacíos del Estado"},
    {"indicador": "Prioridad asignada a la IA",
     "resultado": "Cerca del 50 % alta, 30 % media, 20 % baja",
     "implicacion": "La adopción debe responder a necesidades y confianza, no imponerse"},
    {"indicador": "Privacidad",
     "resultado": "68,3 % asignó máxima importancia",
     "implicacion": "La gobernanza de datos es condición de legitimidad"},
]

@mcp.tool
def consultar_hoja_de_ruta(etapa: int = 0) -> dict:
    """Devuelve la hoja de ruta pública de IA accesible (7 etapas, Tabla 3 del capítulo).
    Con etapa=0 devuelve todas; con etapa=1..7 devuelve esa etapa con su acción y producto verificable."""
    if etapa == 0:
        return {"fuente": "Tabla 3 — Paredes Rojas (Uniandes–BID)",
                "etapas": [{"n": k, **v} for k, v in HOJA_DE_RUTA.items()]}
    if etapa not in HOJA_DE_RUTA:
        return {"error": f"La hoja de ruta tiene 7 etapas; no existe la etapa {etapa}."}
    return {"fuente": "Tabla 3 — Paredes Rojas (Uniandes–BID)",
            "n": etapa, **HOJA_DE_RUTA[etapa]}

@mcp.tool
def consultar_indicadores(tema: str = "") -> dict:
    """Devuelve los indicadores empíricos del capítulo (encuesta a 170 participantes, Tabla 2).
    `tema` filtra por palabra clave (ej: 'privacidad', 'rotación', 'barrera'); vacío devuelve todos."""
    tema_l = (tema or "").lower().strip()
    filtrados = [i for i in INDICADORES
                 if not tema_l or tema_l in (i["indicador"] + i["resultado"] + i["implicacion"]).lower()]
    return {"fuente": "Tabla 2 — encuesta a 170 participantes (Paredes Rojas, Uniandes–BID)",
            "cantidad": len(filtrados), "indicadores": filtrados}

@mcp.tool
def evaluar_iniciativa(descripcion: str, etapas_cubiertas: list[int]) -> dict:
    """Evalúa una iniciativa pública de IA accesible contra la hoja de ruta de 7 etapas.
    `etapas_cubiertas` = etapas (1..7) que la iniciativa ya contempla. Devuelve las
    etapas faltantes con su producto verificable pendiente y observaciones de gobernanza."""
    cubiertas = sorted({e for e in etapas_cubiertas if e in HOJA_DE_RUTA})
    faltantes = [{"n": k, **v} for k, v in HOJA_DE_RUTA.items() if k not in cubiertas]
    observaciones = []
    if 1 not in cubiertas:
        observaciones.append("Sin Diagnóstico no hay línea de base: cualquier evaluación posterior queda sin punto de comparación.")
    if 2 not in cubiertas:
        observaciones.append("Sin Cocreación con la comunidad sorda se repite el patrón que el capítulo critica: tecnología impuesta sin requisitos de derechos.")
    if 3 not in cubiertas:
        observaciones.append("Sin Estándares de privacidad se ignora que el 68,3 % de los encuestados asignó máxima importancia a la privacidad.")
    if 7 not in cubiertas:
        observaciones.append("Sin Evaluación y rendición de cuentas no hay tablero público: la legitimidad queda comprometida.")
    return {"iniciativa": descripcion[:200],
            "etapas_cubiertas": cubiertas,
            "etapas_faltantes": faltantes,
            "observaciones": observaciones,
            "criterio": "Hoja de ruta de 7 etapas — Paredes Rojas (Uniandes–BID)"}

@mcp.tool
def crear_borrador_policy_brief(destinatario: str, tema: str) -> dict:
    """Crea un BORRADOR de policy brief sobre IA accesible para un organismo público.
    NO se envía: requiere revisión y aprobación humana (principio de gobernanza del capítulo)."""
    return {
        "tipo": "borrador", "requiere_aprobacion_humana": True,
        "destinatario": destinatario, "tema": tema,
        "estructura_sugerida": [
            "1. Problema: la disponibilidad técnica no garantiza igualdad de oportunidades.",
            "2. Evidencia: 70 % usa tecnologías de apoyo pero 33,3 % señala la falta de adaptaciones como principal barrera; 56,7 % de rotación laboral en cinco años.",
            "3. Propuesta: adoptar la hoja de ruta de 7 etapas (Diagnóstico → Evaluación y rendición de cuentas).",
            "4. Presupuesto y gobernanza: financiamiento del escalamiento + tablero público.",
        ],
        "advertencia": "Borrador generado por IA sobre evidencia publicada. Debe revisarlo una persona antes de circular.",
    }

print("🔌 MCP 'observatorio-ia-accesible' listo — 4 tools sobre datos reales del capítulo")
```

```python
# ============================================================
# 🔌 §4.2 · Probar el catálogo con el cliente MCP oficial
# ============================================================
# QUÉ HACE: abre un cliente MCP contra el server (transporte
# in-memory, mismo protocolo que por red) y lista las tools con
# sus schemas — el "contrato" que después consume el agente.
# ------------------------------------------------------------
import asyncio, json
import nest_asyncio
nest_asyncio.apply()

from fastmcp import Client

async def _catalogo():
    async with Client(mcp) as c:
        return await c.list_tools()

MCP_TOOLS_INFO = asyncio.get_event_loop().run_until_complete(_catalogo())
print(f"🧰 El observatorio expone {len(MCP_TOOLS_INFO)} tools MCP:")
for t in MCP_TOOLS_INFO:
    print(f"   • {t.name}: {(t.description or '').splitlines()[0][:78]}")
```

```python
# ============================================================
# 🔌 §4.3 · Tools MCP → StructuredTools de LangChain (dinámico)
# ============================================================
# QUÉ HACE: por cada tool del server crea un puente sync→async
# y la registra como herramienta de LangChain con el MISMO schema
# JSON que declara el server. El agente las ve como cualquier
# otra tool; por debajo, cada llamada viaja por el protocolo MCP.
# ------------------------------------------------------------
from langchain_core.tools import StructuredTool

def _llamar_mcp_sync(tool_name: str, **kwargs) -> str:
    async def _run():
        async with Client(mcp) as c:
            res = await c.call_tool(tool_name, kwargs)
            return "\n".join(getattr(b, "text", str(b)) for b in res.content)
    return asyncio.get_event_loop().run_until_complete(_run())

def _crear_tool_langchain(info):
    nombre = info.name
    def _runner(**kwargs):
        return _llamar_mcp_sync(nombre, **kwargs)
    return StructuredTool.from_function(
        func=_runner,
        name=f"mcp_{nombre}",
        description=f"[MCP·observatorio] {info.description or nombre}",
        args_schema=info.inputSchema,
    )

mcp_tools = [_crear_tool_langchain(t) for t in MCP_TOOLS_INFO]
print("🔗 Tools MCP para el agente:", [t.name for t in mcp_tools])

# prueba directa del puente
print("\nprueba →", mcp_tools[0].invoke({"etapa": 3})[:160], "…")
```

## §5 · Tools propias

**🎯 Objetivo:** completar el arsenal del agente con herramientas nuestras. La clave es
`consultar_capitulo`: **conecta el RAG de la §3 al agente**, para que el bot decida solo
cuándo buscar en la evidencia.

**⚙️ Qué hace el bloque:** define tres tools con el decorador `@tool` de LangChain —
búsqueda en el capítulo (RAG), calculadora segura (para proyecciones rápidas, ej. aplicar
un porcentaje del capítulo a la población de una provincia) y fecha/hora — y arma la
lista final `TOOLS` combinándolas con las del MCP.

```python
# ============================================================
# 🛠️ §5 · Tools propias + arsenal completo
# ============================================================
from langchain_core.tools import tool
from datetime import datetime
from zoneinfo import ZoneInfo

@tool
def consultar_capitulo(pregunta: str) -> str:
    """Busca en el capítulo Uniandes–BID (RAG sobre la base vectorial persistente).
    Usala para preguntas sobre la investigación: metodología, hallazgos, marco teórico,
    citas, contexto argentino, conclusiones."""
    r = responder_con_fuentes(pregunta)
    return f"{r['respuesta']}\n[fuentes: {', '.join(r['fuentes'])}]"

@tool
def calculadora(expresion: str) -> str:
    """Evalúa una expresión aritmética simple, ej: '0.333 * 45000' para proyectar
    un porcentaje del capítulo sobre una población objetivo."""
    import re
    if not re.fullmatch(r"[0-9+\-*/().,\s]+", expresion):
        return "Expresión no permitida: solo números y + - * / ( )."
    try:
        return str(eval(expresion, {"__builtins__": {}}, {}))
    except Exception as e:
        return f"Error al evaluar: {e}"

@tool
def fecha_hora_actual() -> str:
    """Fecha y hora actual en Argentina (para fechar informes y borradores)."""
    return datetime.now(ZoneInfo("America/Argentina/Buenos_Aires")).strftime("%A %d/%m/%Y, %H:%M hs")

tools_propias = [consultar_capitulo, calculadora, fecha_hora_actual]
TOOLS = tools_propias + mcp_tools
print(f"🛠️ Arsenal del agente: {len(TOOLS)} tools →", [t.name for t in TOOLS])
```

## §6 · Agente ReAct con memoria

**🎯 Objetivo:** el corazón del sistema. El agente recibe la pregunta del funcionario
y **decide solo** qué combinación de herramientas usar: evidencia del capítulo (RAG),
hoja de ruta o indicadores (MCP), cálculo, o respuesta directa.

**⚙️ Qué hace el bloque:** crea el agente con import a prueba de versiones (API nueva
de LangChain v1 o la clásica de LangGraph), le pasa el *system prompt* con las reglas de
uso de cada herramienta, y le enchufa un **checkpointer** — la memoria por `thread_id`.
Después, `hablar()` empaqueta un turno completo y extrae qué tools se usaron (eso alimenta
los chips de la interfaz).

**🗣️ Al presentar:** "yo no programo el flujo: describo las herramientas y el agente razona
qué necesita. Miren la demo: tres preguntas distintas disparan tres rutas distintas".

```python
# ============================================================
# 🤖 §6 · Agente + checkpointer
# ============================================================
from langgraph.checkpoint.memory import MemorySaver

try:
    from langchain.agents import create_agent as _crear_agente
    _API = "langchain.create_agent"
except Exception:
    from langgraph.prebuilt import create_react_agent as _crear_agente
    _API = "langgraph.create_react_agent"

SYSTEM_CHATBOT = """Sos un asesor de políticas públicas especializado en IA accesible e
inclusión laboral de personas con discapacidad auditiva. Tu conocimiento proviene del
capítulo Uniandes–BID de Paredes Rojas (investigación real: encuesta a 170 participantes,
50 narrativas laborales y entrevistas institucionales en Argentina).

Herramientas y cuándo usarlas:
- consultar_capitulo → preguntas sobre la investigación (metodología, hallazgos, marco, conclusiones).
- mcp_consultar_hoja_de_ruta → las 7 etapas de implementación y sus productos verificables.
- mcp_consultar_indicadores → porcentajes y resultados empíricos (uso de tecnología, rotación, privacidad…).
- mcp_evaluar_iniciativa → cuando el usuario describe SU proyecto y quiere saber qué le falta.
- mcp_crear_borrador_policy_brief → cuando piden un documento para un organismo; SIEMPRE aclarar que es borrador y requiere aprobación humana.
- calculadora → proyecciones numéricas puntuales.
- fecha_hora_actual → fechar informes.

Reglas:
- Citá porcentajes exactos del capítulo cuando existan; no inventes datos.
- Si algo no está en la evidencia, decilo con honestidad.
- Español rioplatense, profesional, claro, orientado a la acción pública.
"""

checkpointer = MemorySaver()

if _API == "langchain.create_agent":
    agente = _crear_agente(model=llm, tools=TOOLS,
                           system_prompt=SYSTEM_CHATBOT, checkpointer=checkpointer)
else:
    agente = _crear_agente(model=llm, tools=TOOLS,
                           prompt=SYSTEM_CHATBOT, checkpointer=checkpointer)

print(f"✅ Agente listo ({_API}) — {len(TOOLS)} tools, memoria por thread_id")
```

```python
# ============================================================
# 🤖 §6.1 · hablar(): un turno completo + tools usadas
# ============================================================
from langchain_core.messages import HumanMessage, AIMessage

def hablar(mensaje: str, session_id: str = "demo") -> dict:
    config = {"configurable": {"thread_id": session_id}}
    resultado = agente.invoke({"messages": [HumanMessage(content=mensaje)]}, config=config)

    pasos, respuesta = [], ""
    for m in resultado["messages"]:
        if isinstance(m, AIMessage) and getattr(m, "tool_calls", None):
            pasos += [tc["name"] for tc in m.tool_calls]
        if isinstance(m, AIMessage) and m.content:
            respuesta = m.content if isinstance(m.content, str) else str(m.content)
    return {"respuesta": respuesta, "tools_usadas": pasos}

# --- demo: tres consultas típicas de un decisor público ---
for pregunta in [
    "¿Qué porcentaje priorizó la privacidad y qué implica para la política de datos?",  # → MCP indicadores
    "Tengo un proyecto provincial: compramos tablets con subtitulado y capacitamos al personal. ¿Qué etapas de la hoja de ruta me faltan?",  # → MCP evaluar (cubre ~4 y 5)
    "¿Qué organizaciones fueron entrevistadas y qué rol cumplen las ONG según la evidencia?",  # → RAG capítulo
]:
    r = hablar(pregunta, session_id="demo-agente")
    print("🧑 ", pregunta[:90])
    print("🤖 ", r["respuesta"][:300])
    print("🛠️ ", r["tools_usadas"] or "(sin tools)"); print("-"*72)
```

## §9 · Interfaz del asesor (patrón de la Unidad 4)

**⚙️ Qué hacen los bloques:**
1. **§9.1** — Registra el callback Python (`register_callback`) que la interfaz invoca.
   Devuelve `IPython.display.JSON` (objeto nativo) para evitar problemas de escapado
   con acentos, y recorta las tools del turno actual del historial acumulado del thread.
2. **§9.2** — Renderiza la interfaz HTML glass. El campo `thread_id` de arriba permite

**🗣️ Al presentar:** "los chips de abajo de cada respuesta son la prueba visual de la

```python
# ============================================================
# 💬 §9.1 · Callback Python del chat
# ============================================================
from IPython.display import JSON

_TURNOS_PREVIOS: dict[str, int] = {}

def _cb_chat(mensaje: str, session_id: str):
    try:
        mensaje = (mensaje or "").strip()
        if not mensaje:
            return JSON({"respuesta": "Escribime tu consulta 🙂", "tools": []})

        r = hablar(mensaje, session_id=session_id or "organismo-1")

        previas = _TURNOS_PREVIOS.get(session_id, 0)
        tools_turno = r["tools_usadas"][previas:]
        _TURNOS_PREVIOS[session_id] = len(r["tools_usadas"])

        return JSON({"respuesta": r["respuesta"], "tools": tools_turno, "session": session_id})
    except Exception as e:
        import traceback; print(traceback.format_exc())
        return JSON({"respuesta": f"Error interno: {str(e)[:150]}", "tools": [], "error": True})

try:
    from google.colab import output as _co
    _co.register_callback("asesor.chat", _cb_chat)
    print("✅ Callback 'asesor.chat' registrado — ejecutá la celda siguiente para abrir la interfaz.")
except ImportError:
    print("⚠️  Fuera de Colab: la interfaz no podrá invocar al kernel.")
```
